```python
import pandas
from matplotlib import pyplot
import numpy as np

file = pandas.read_csv("abalone.csv", index_col=0)
```

```python
age = file.loc[:, 'age'].to_numpy()
coclender = file.loc[:, 'height'].to_numpy()

pyplot.plot(age, coclender, 'bo')
```

```python
coclender = file.loc[:, 'height'].to_numpy()

mn, mx = 1e20, 0
for fut, e in enumerate(coclender):
    if e == 0.0:
        print(fut)

    mn = min(mn, e)
    mx = max(mx, e)

print(mn, mx)
```

```python
coclender = file.loc[:, 'height'].to_numpy()

for fut, e in enumerate(coclender):
    if e > 0.4:
        print(fut)

print(mn, mx)
```
